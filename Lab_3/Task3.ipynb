```python
#Write a python code that solves the Rook’s problem for 10 rooks using evolutionary algorithms. Explain in 
#your  code  how  you  represented  the data,  and how  you  mapped different  algorithm  steps  to  model  your 
#problem. 

#konfiguration
N=10 # storleken på schackbrädet (10x10) och antalet torn som ska placeras
N_rooks=N # antalet torn
N_population=100 # antalet rutor på brädet
N_generations=1000 # max antal generationer
mutation_rate=0.2 # Mutation rate



```

```python
# steg 1:  Data representation:
# Vi representerar varje lösning som en lista av längd N, där indexet representerar kolumnen och värdet representerar raden där tornet är placerat. Till exempel, [0, 2, 4, 6, 8, 1, 3, 5, 7, 9] betyder att det finns torn i kolumnerna 0-9 och de är placerade.
# Exempel: [0, 1, 2...] betyder torn på (0,0), (1,1), (2,2).
# Detta garanterar automatiskt att inga torn delar rad eller kolumn om vi använder unika tal.

import random
import numpy as np

def create_individual():
    # Skapa en individ genom att slumpmässigt placera N torn i unika rader
    genes = list(range(N))  # Starta med en lösning där varje torn är i sin egen rad
    random.shuffle(genes)  # Blanda för att skapa en slumpmässig lösning
    return genes
```

```python
#---Fitness-funktion:---
# Fitness-funktionen räknar antalet par av torn som attackerar varandra. Eftersom vi representerar lösningen som en permutation av rader, kommer det inte att finnas några attacker i rader eller kolumner. Vi behöver bara kontrollera diagonalerna.

def fitness(individual):
    conflicts = 0   # Räkna antalet attackerande par
    for i in range(N):
        for j in range(i + 1, N):
            # Kontrollera om torn i kolumnerna i och j attackerar varandra diagonalt
            if abs(individual[i] - individual[j]) == abs(i - j):
                conflicts += 1  # Öka konflikterna om de attackerar varandra
    return conflicts
```

```python
#---steg 3: välja bästa individer:

def selection(population):
    # Välj de bästa individerna baserat på fitness (lägre är bättre)
    sort_pop= sorted(population, key=fitness)  # Sortera populationen efter fitness
    return sort_pop[:N_population // 2]  # Behåll de bästa hälften, 50%
```

```python
# steg 4: Crossover:
def crossover(p1, p2):
    i= random.randint(1,N-2)  # Välj en crossover-punkt
    child1= p1[:i] + [gene for gene in p2 if gene not in p1[:i]]  # Skapa barn genom att kombinera föräldrar
    child2= p2[:i] + [gene for gene in p1 if gene not in p2[:i]]


    return child1, child2


```

```python
# steg 5: Mutation:

def mutate(individual):
     # Skapa en kopia för att inte ändra på föräldern i 'selected'
    new_ind = individual.copy() 
    if random.random() < mutation_rate:
        i, j = random.sample(range(N), 2)  # Välj två positioner att byta
        new_ind[i], new_ind[j] = new_ind[j], new_ind[i]  # Byt plats på två torn
    return new_ind

```

```python
# Evolutionary algorithm main loop:
population= [create_individual() for _ in range(N_population)]  # Skapa initial population

print("Startar evolutionär algoritm för Rook's problem...")

for generation in range(N_generations):
    #1. Utvärdera fitness
    population.sort(key=fitness)  # Sortera populationen efter fitness
    best_induvidual = population[0]  # Den bästa individen är den med lägst fitness
    best_fitness = fitness(best_induvidual)
    print(f"Generation {generation}: Bästa fitness = {best_fitness}:Bästa individ = {best_induvidual}")

    if best_fitness == 0:  # Om vi har hittat en lösning utan konflikter
        print("Lösning hittad!")
        print(population[0])  # Skriv ut den bästa lösningen
        break

    #2. Välj de bästa individerna
    selected = selection(population)
    #3. Skapa nästa generation genom crossover
    next_generation=[best_induvidual]  # Behåll den bästa individen i nästa generation

    while len(next_generation) < N_population:
        #väljer två föräldrar slumpmässigt från de valda individerna
        p1,p2=random.sample(selected,2)
        #skapa två avkommor genom crossover
        os_1,os_2=crossover(p1,p2)
        #mutera avkommorna
        os_1=mutate(os_1)
        os_2=mutate(os_2)

        next_generation.append(os_1)
        if len(next_generation) < N_population:
            next_generation.append(os_2)
        #fittness avkommorna
             #fittness av avkommorna, max fit
        # max_fit=max(fitness(os_1),fitness(os_2))
        # #print(max_fit)
        
        # fitness_os_1=fitness(os_1)
        # fitness_os_2=fitness(os_2)
        # #lägg till avkommorna i nästa generation
        # next_generation.extend([os_1,os_2])
    population=next_generation[:N_population]  # Begränsa populationen till N_population

    
```
