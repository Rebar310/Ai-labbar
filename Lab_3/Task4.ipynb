```python
#Write a python code that solves the following Knapsack1 problem. Make sure your implementation allows 
#for more than one selection of each item as a solution. Evolva algortima (EA) is a good choice for this problem.
import random
#steg 1: Data representation
capacity = 15
weights= [1, 1, 2, 4, 12]
values = [1, 2, 2, 10, 4]

#max antal av varje item som kan väljas
max_count=[capacity//w for w in weights]

n_init=50
# --- Parametrar ---
N_population = 50
N_generations = 100
mutation_rate = 0.1

def create_individual():
    return [random.randint(0, max_count[i]) for i in range(len(values))]


```

```python
# steg 2: skapar individer och population

import numpy as np

#skapa population med slumpmässiga antal av varje item, upp till max_count
pop_init= np.array([[random.randint(0, max_count[i]) for i in range(len(values))] for _ in range(n_init)])

# def initialize_population(pop_size):
#     population = []
#     for _ in range(pop_size):
#         individual = [random.randint(0, max_count[i]) for i in range(len(values))]
#         population.append(individual)
#     return population

print(pop_init)
```

```python
# steg 3: fitnessfunktion, beräknar värdet av en individ
def fitness(population):
    ind=np.array(population)
    w_sum= np.sum(ind * weights)
    v_sum= np.sum(ind * values)
    # straffa individer som överskrider kapaciteten
    if w_sum > capacity:
        return 0
    return v_sum
#printa svaren för kontroll
```

```python
# steg 4: selektion, väljer individer baserat på fitness

def selection(population):
   selected= random.sample(population, 3)
   return max(selected, key=fitness)
```

```python
#steg 5: crossover, kombinerar två föräldrar för att skapa avkomma
def crossover(parent1, parent2):
    i = random.randint(1, len(parent1) - 1)
    child1 = parent1[:i] + parent2[i:]
    child2 = parent2[:i] + parent1[i:]
    return child1, child2
```

```python
#steg 6: mutation, introducerar variation i avkomman

def mutate(individual, mutation_rate=0.1):
    for i in range(len(individual)):
        if random.random() < mutation_rate:
            individual[i] = random.randint(0, max_count[i])
    return individual
```

```python
# Evolutionary algorithm main loop:
population= [create_individual() for _ in range(N_population)]  # Skapa initial population

print("Startar evolutionär algoritm för knapsack problem...")

for generation in range(N_generations):
    #1. Utvärdera fitness
    population.sort(key=fitness, reverse=True)  # Sortera populationen efter fitness
    best_individual = population[0]
    best_fitness = fitness(best_individual)

    if generation % 10 == 0:
        print(f"Generation {generation}: Bästa värde = {best_fitness}, Vikt = {sum(np.array(population[0])*weights)}")

      # Skapa nästa generation
    next_generation = [best_individual] # Elitism: behåll den bästa individen

    while len(next_generation) < N_population:
        #välj två föräldrar baserat på fitness
        p1=selection(population)
        p2=selection(population)
        #skapa två avkommor genom crossover
        c1,c2=crossover(p1,p2)

        #mutera avkomman
        next_generation.append(mutate(c1))
        if len(next_generation) < N_population:
            next_generation.append(mutate(c2))
     
    population=next_generation # Begränsa populationen till N_population

    
```

```python
#slutresultat
population.sort(key=fitness, reverse=True)
best = population[0]
print("Bästa lösning:", best)
print("Bästa värde:", fitness(best))
print("Bästa vikt:", sum(np.array(best) * weights))
```
