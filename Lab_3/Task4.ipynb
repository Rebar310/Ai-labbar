```python
#Write a python code that solves the following Knapsack1 problem. Make sure your implementation allows 
#for more than one selection of each item as a solution. Evolva algortima (EA) is a good choice for this problem.

#steg 1: Data representation
capacity = 15
weights= [1, 1, 2, 4, 12]
values = [1, 2, 2, 10, 4]

#max antal av varje item som kan väljas
max_count=[capacity//w for w in weights]

n_init=50

```

```python
# steg 2: skapar individer och population
import random
import numpy as np

#skapa population med slumpmässiga antal av varje item, upp till max_count
pop_init= np.array([[random.randint(0, max_count[i]) for i in range(len(values))] for _ in range(n_init)])

# def initialize_population(pop_size):
#     population = []
#     for _ in range(pop_size):
#         individual = [random.randint(0, max_count[i]) for i in range(len(values))]
#         population.append(individual)
#     return population

print(pop_init)
```

```python
# steg 3: fitnessfunktion, beräknar värdet av en individ
def fitness(population):
    w_sum= (population * weights).sum(axis=1)
    v_sum= (population * values).sum(axis=1)
    # straffa individer som överskrider kapaciteten
    penalty = np.where(w_sum > capacity, 0, v_sum)
    return penalty
```

```python
# steg 4: selektion, väljer individer baserat på fitness

def selection(population):
   contestaants = random.sample(population, 3)
   return max(contestaants, key=fitness)
```

```python
#steg 5: crossover, kombinerar två föräldrar för att skapa avkomma
def crossover(parent1, parent2):
    i = random.randint(1, len(parent1) - 1)
    child1 = parent1[:i] + parent2[i:]
    child2 = parent2[:i] + parent1[i:]
    return child1, child2
```

```python
#steg 6: mutation, introducerar variation i avkomman

def mutate(individual, mutation_rate=0.1):
    for i in range(len(individual)):
        if random.random() < mutation_rate:
            individual[i] = random.randint(0, max_count[i])
    return individual
```
